**A company started to invest in digital marketing as a new way of their product
promotions. For that they collected data and decided to carry out a study on it.**

```python
import pandas as pd
import numpy as np
sales_add = pd.read_csv('/content/Sales_add.csv')
sales_add.head()
```

**● The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.**

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(sales_add['Sales_before_digital_add(in $)'], sales_add['Sales_After_digital_add(in $)'])
print('Here we take, \n\033[1mH0: There is no increase in sales after stepping into digital marketing\nH1: There is increase in sales after stepping into digital marketing')
```

```python
if p_value< 0.05:
  print('P value = ',p_value, '\n\033[1mNull hypothesis rejected')
  print('i.e. There is increase in sales after stepping into digital marketing')
else:
  print('P value = ',p_value, '\n\033[1mNull hypothesis accepted')
  print('i.e. There is no increase in sales after stepping into digital marketing')
```

**● The company needs to check whether there is any dependency between
the features “Region” and “Manager”.**

```python
cont_tab = pd.crosstab(sales_add['Region'], sales_add['Manager'])
print('Here we take, \n\033[1mH0: There is no dependency between the features “Region” and “Manager”.\nH1: There is dependency between the features “Region” and “Manager”.\n')
cont_tab
```

```python
#test of independence
from scipy.stats import chi2_contingency
chi_stat, p_val, dof, exp_f = chi2_contingency(cont_tab)

if p_val< 0.05:
  print('P value = ',p_val, '\n\033[1mNull hypothesis rejected')
  print('i.e. There is dependency between the features “Region” and “Manager”.')
else:
  print('P value = ',p_val, '\n\033[1mNull hypothesis accepted')
  print('i.e. There is no dependency between the features “Region” and “Manager”.')
```
