**5. Perform some meaningful visualizations using the dataset.**

**i. Use of bar chart for visualizing SL values of each Iris classification**

```python
import matplotlib.pyplot as plt
plt.bar(iris_Data['Classification'],iris_Data['SL'], color = 'Green', width=0.4)
plt.title('SL for each Iris classification')
plt.xticks(rotation = 45)
plt.xlabel('Iris classification')
plt.ylabel('SL')
```

**ii. Use of subplots in matplotlib for variability check of comparable parameters**

```python
s, twoDplot = plt.subplots(2,2)
s.tight_layout(pad = 3.5)
s.suptitle('Comparison of SW,SL,PW,PL using subplots')
twoDplot[0,0].plot(iris_Data['Classification'], iris_Data['SW'])
twoDplot[0,0].set_title('SW')
twoDplot[0,1].plot(iris_Data['Classification'], iris_Data['SL'])
twoDplot[0,1].set_title('SL')
twoDplot[1,0].plot(iris_Data['Classification'], iris_Data['PW'])
twoDplot[1,0].set_title('PW')
twoDplot[1,1].plot(iris_Data['Classification'], iris_Data['PL'])
twoDplot[1,1].set_title('PL')
```

**iii. Scatter plot to view SW-SL and PW-PL values**

```python
sub_data = iris_Data['Classification']
plt.scatter(iris_Data['SL'],iris_Data['SW'], color = 'Red', marker='*')
plt.scatter(iris_Data['PL'],iris_Data['PW'], color = 'Blue', marker='^')
plt.xlabel('SL/PL')
plt.ylabel('SW/PW')
```

**iv. Boxplot to see the variability of SL, SW, PL, PW values**

```python
plt.boxplot(iris_Data[['SL','SW','PL','PW']])
plt.xticks([1,2,3,4],['SL','SW','PL','PW'])
plt.ylabel('Value range')
plt.title('Boxplot of SL,SW,PL and PW')
```
