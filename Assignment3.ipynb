**2. What do you think about the students' parental level of education?**

```python
import statistics as stat
print('By taking the mode we can infer that most of the parental level of education is: ',stat.mode(studentdata['parental level of education']))
print('\nDetailed value counts are:\n',studentdata['parental level of education'].value_counts())
```

**3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course**

```python
gender_math = studentdata.groupby('gender')['math score'].mean()
print(gender_math, '\n\033[1;37;35m Highest on average for math: ',gender_math.max(),' by gender: ',gender_math.idxmax(),'\033[0m \n')

gender_reading = studentdata.groupby('gender')['reading score'].mean()
print(gender_reading,'\n\033[1;37;35m Highest on average for reading: ',gender_reading.max(),' by gender: ',gender_reading.idxmax(),'\033[0m \n')

gender_writing = studentdata.groupby('gender')['writing score'].mean()
print(gender_writing,'\n\033[1;37;35m Highest on average for writing: ',gender_writing.max(),' by gender: ',gender_writing.idxmax(),'\033[0m \n')


course_math = studentdata.groupby('test preparation course')['math score'].mean()
print(course_math, '\n\033[1;37;35m Highest on average for math: ',course_math.max(),' with test preparation course: ',course_math.idxmax(),'\033[0m \n')

course_reading= studentdata.groupby('test preparation course')['reading score'].mean()
print(course_reading, '\n\033[1;37;35m Highest on average for reading: ',course_reading.max(),' with test preparation course: ',course_reading.idxmax(),'\033[0m \n')

course_writing = studentdata.groupby('test preparation course')['writing score'].mean()
print(course_writing, '\n\033[1;37;35m Highest on average for writing: ',course_writing.max(),' with test preparation course: ',course_writing.idxmax(),'\033[0m \n')
```

**4. What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course**


```python
gender_mathvar = studentdata.groupby('gender')['math score'].var()
print(gender_mathvar, '\n\033[1;33;35m Highest scoring variation for math: ',gender_mathvar.max(),' by gender: ',gender_mathvar.idxmax(),'\033[0m \n')

gender_readingvar = studentdata.groupby('gender')['reading score'].var()
print(gender_readingvar,'\n\033[1;33;35m Highest scoring variation for reading: ',gender_readingvar.max(),' by gender: ',gender_readingvar.idxmax(),'\033[0m \n')

gender_writingvar = studentdata.groupby('gender')['writing score'].var()
print(gender_writingvar,'\n\033[1;33;35m Highest scoring variation for writing: ',gender_writingvar.max(),' by gender: ',gender_writingvar.idxmax(),'\033[0m \n')


course_mathvar = studentdata.groupby('test preparation course')['math score'].var()
print(course_mathvar, '\n\033[1;33;35m Highest scoring variation for math: ',course_mathvar.max(),' with test preparation course: ',course_mathvar.idxmax(),'\033[0m \n')

course_readingvar= studentdata.groupby('test preparation course')['reading score'].var()
print(course_readingvar, '\n\033[1;33;35m Highest scoring variation for reading: ',course_readingvar.max(),' with test preparation course: ',course_readingvar.idxmax(),'\033[0m \n')

course_writingvar = studentdata.groupby('test preparation course')['writing score'].var()
print(course_writingvar, '\n\033[1;33;35m Highest scoring variation for writing: ',course_writingvar.max(),' with test preparation course: ',course_writingvar.idxmax(),'\033[0m \n')
```
